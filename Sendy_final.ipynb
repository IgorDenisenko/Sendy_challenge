# Features

```python
#Time features
def add_time_feature(data):
    """
    функция и для трейна и для теста
    """
    # через сколько гонщик подтвердил, что он приедет за заказом
    data['t_from_pl_to_conf_1_2'] = (pd.to_datetime(data["Confirmation - Time"]) - pd.to_datetime(data["Placement - Time"])).astype('timedelta64[s]')
    # через сколько времени после подтверждения гонщик приехал, чтобы забрать заказ
    data['t_from_conf_to_arrival_2_3'] = (pd.to_datetime(data["Arrival at Pickup - Time"]) - pd.to_datetime(data["Confirmation - Time"])).astype('timedelta64[s]')
    # промежуток времени между  тем когда гонщик приехал забрать и заказ, и чеез сколько он его реально забрал
    data['t_from_arrival_to_pickup_3_4'] = (pd.to_datetime(data["Pickup - Time"]) - pd.to_datetime(data["Arrival at Pickup - Time"])).astype('timedelta64[s]')
    
    #Суммы по промежуткам времени
    
    # сколько времени прошло между размещением заказа и временем его приезда за заказом
    data['time_interval_1_3'] = data['t_from_pl_to_conf_1_2'] + data['t_from_conf_to_arrival_2_3']
    
    # сколько времени прошло между подтверждением заказа и временем,когда он забрал заказ
    data['time_interval_2_4'] = data['t_from_conf_to_arrival_2_3'] + data['t_from_arrival_to_pickup_3_4']
    
    # сколько времени прошло между размещением заказа и временем,когда он забрал заказ
    data['time_interval_1_4'] = data['t_from_pl_to_conf_1_2']+data['t_from_conf_to_arrival_2_3']+data['t_from_arrival_to_pickup_3_4']
    
    
    
    data["Pickup - hour"] = pd.to_datetime(data["Pickup - Time"]).dt.hour.astype("category")
    # является ли выходным
    data['is_holiday'] = np.where(data['Pickup - Weekday (Mo = 1)'] >=6,1,0)
    
#Speed features
def add_speed_feature(data_train,thresh_speed=55):
    """
    Только для трейна
    """
    data_train['speed'] = data_train['Distance (KM)'] / (data_train['Time from Pickup to Arrival'] / 3600) 
    data_train['is_anomaly'] = np.where(data_train['speed'] >=thresh_speed,1,0)
    
#Coordinates features
def concat_coordinates(any_data):
    """
    как для трейна, так и для теста
    """
    def plus(x):
        return str(x[0])+ ' ' + str(x[1])
    
    any_data['lat_lon_pickup'] = (any_data[['Pickup Lat','Pickup Long']]).apply(lambda x: plus(x),axis=1)
    any_data['lat_lon_arrival'] = (any_data[['Destination Lat','Destination Long']]).apply(lambda x: plus(x),axis=1) 
    
def fill_column_with_mean(any_data, column_name_str):
    # fill temperature with mean
    mean = any_data[column_name_str].mean()
    any_data[column_name_str].fillna(value=mean,inplace=True)
```

```python
# Конкатенация категориальых переменных

def genetare_list_columns_to_concat(list_cat_colums,count_of_comb):
    """    
    list_cat_colums - лист с категориальными переменными,которые нужно состакать
    count_of_comb - количество комбинаций (состакать по два, по три и тд)
    """
    final_l = []
    for i in list(it.combinations(list_cat_colums,r=count_of_comb)): 
        final_l.append(list(i))
    return final_l

def generate_new_columns(any_data,list_of_stacked_colums):
    """
    list_of_stacked_colums - лист с комбинациями столбцов
    """
    
    for i in list_of_stacked_colums:
        print(i)
        name = ''
        for name_column in i:
            name+=name_column+'_'
        print(name)
        any_data[name[:-1]] = ''
        for name_column in i: 
            any_data[name[:-1]] += any_data[name_column]
        name=''

def create_genaretad_stack_columns(any_data,cat_columns,count_combination):
    """
    """
    full_list_columns = []
    for k in range(2,count_combination+1):
        full_list_columns.append(genetare_list_columns_to_concat(cat_columns,k))
    for i in full_list_columns:
        generate_new_columns(any_data,i)
```

# Target encode with validation

```python
def smoothed_likelihood (mean_target, n_rows, global_median, alpha):
  """Расчет "сглаженного" среднего"
  return (mean_target*n_rows + global_median*alpha) / (n_rows+alpha)
```

```python
def mean_encode_data (X_train, X_test,feature_name = 'rider_av_speed_1', groupby_name = 'Rider Id',
                                    mean_name = 'speed',alpha = 0.2, use_anomaly = False):
    """
    Заполнить фичу в тесте, рассчитанную по трейну
    -------------------------------------------
    feature_name - название созданной фичи - mean_encoding существующей фичи
    groupby_name - по какую фичу кодируем (например Rider id)
    mean_name - по какой фиче получаем mean_encoding (например, speed)(target)
    use_anomaly = False - не учитываем аномалии,  True - учитываем
    for_test = False - расчет идет для валидации, True - для тестовой части финального теста
    """
    if use_anomaly:
        # по валидационным фолдам считаем  средние и сколько раз встречается  райдер
        rider_dict = dict(X_train.groupby(by = groupby_name)[mean_name].mean())
        #mean non ano speed global by fold
        global_mean_by_val_folds = X_train[mean_name].mean()
        # количество строк в фолде
        num_rows = dict(X_train.groupby(groupby_name)[mean_name].count())

    else:
        rider_dict = dict(X_train[X_train['is_anomaly'] == 0].groupby(by = groupby_name)[mean_name].mean())
        global_mean_by_val_folds = X_train[X_train['is_anomaly'] == 0][mean_name].mean()
        # количество райдеров в фолде
        num_rows = dict(X_train[X_train['is_anomaly'] == 0].groupby(groupby_name)[mean_name].count())

    # заполним валидационный фолд / тест обычными средними значениями с трейновых фолдов
    X_train[feature_name] = 0
    X_test[feature_name] = X_test[groupby_name].map(rider_dict) # 1
    X_test['help_feature'] = X_test[groupby_name].map(num_rows) # можно дропнуть или  заюзать как фичу
    X_test[feature_name]  = X_test[[feature_name,'help_feature']].apply((lambda x: smoothed_likelihood(x[0], x[1],
                                                                                                      global_mean_by_val_folds, 
                                                                                                      alpha)), 
                                                                                                      axis = 1)
    
    X_test[feature_name].fillna(global_mean_by_val_folds, inplace = True)  
        
    return X_train, X_test
```

```python
data = data.set_index('Order No')
test_zindi = test_zindi.set_index('Order No')
X = data.copy()
y = data['Time from Pickup to Arrival'].copy()
test_to_prediction = test_zindi.copy()

X['Distance (KM)'] = X['Distance (KM)'].astype('float64')
test_to_prediction['Distance (KM)'] = test_to_prediction['Distance (KM)'].astype('float64')
```

```python
cat_features_to_learn = ['User Id','Rider Id','lat_lon_pickup','lat_lon_arrival'] + diff_columns #'lat_lon_pickup','lat_lon_arrival' #'is_anomaly'
params_dict={'To_mean_encode':cat_features_to_learn, # какие фичи кодируем таргет
             'By_mean_encode':['speed','t_from_arrival_to_pickup_3_4','is_anomaly'],#что используем в качестве таргета при енкод
             'use_anomaly':[True,False]}# использовать аномалии или нет
```

```python
#model
clf= CatBoostRegressor(iterations=1200, learning_rate = 0.03, max_depth=3,
                  loss_function = 'RMSE', verbose = True, use_best_model = False,
            random_seed= 42,
                task_type = 'GPU')
```

```python
# Основная функци обучение модели + валидация 

MODE = 'validation'
def test_prediction(X_train,y,n_splits = 3,n_splits_sub = 3,params_dict = params_dict,
                    K_alpha=5,mode = MODE,test=test_to_prediction):


    preds_buf = []
    target_train_buf = []
    target_test_buf = []
    err_buf = []
    # mean_encoding 
    num_iter=1
    cv = KFold(n_splits=n_splits, random_state=42, shuffle=True)
    cv_sub = KFold(n_splits=n_splits_sub, random_state=42, shuffle=True)
    combinations = it.product(*(params_dict[Name] for Name in params_dict)) # всевозможные комбинации фичей
    n_col = 0
    for i in list(combinations):
        if (i[1] == 'is_anomaly' and i[2] == False) or (i[1] == 'speed'  and i[2] == True):
            continue
        else:
            n_col +=1

    for train_index, test_index in cv.split(X, y):
        if mode == 'validation':
            X_train,X_test,y_train,y_test = X.iloc[train_index].copy(), \
                                              X.iloc[test_index].copy(), \
                                              y.iloc[train_index].copy(), \
                                              y.iloc[test_index].copy()
        elif mode == 'prediction':
            X_train,X_test,y_train = X.copy(),test.copy(),y.copy()
            
        combinations = it.product(*(params_dict[Name] for Name in params_dict)) # всевозможные комбинации колонок
        make_feat = 0
        for i in list(combinations):
            if (i[1] == 'is_anomaly' and i[2] == False) or (i[1] == 'speed'  and i[2] == True):
                continue
            else:
                X_train,X_test =  mean_encode_data(X_train,
                                                    X_test,
                                                    feature_name = str(i[0]) +'_'+str(i[1]) +'_'+str(i[2]),
                                                    groupby_name = i[0],
                                                    mean_name = i[1],
                                                    alpha = K_alpha,
                                                    use_anomaly = i[2])
                for train_index_sub, test_index_sub in cv_sub.split(X_train,y_train):
                    X_train_sub,X_test_sub,y_train_sub,y_test_sub = X_train.iloc[train_index_sub].copy(), \
                                                                        X_train.iloc[test_index_sub].copy(), \
                                                                        y_train.iloc[train_index_sub].copy(), \
                                                                        y_train.iloc[test_index_sub].copy()
                    X_train_sub,X_test_sub = mean_encode_data(X_train_sub,
                                                              X_test_sub,
                                                              feature_name = str(i[0]) +'_'+str(i[1]) +'_'+str(i[2]),
                                                                groupby_name = i[0],
                                                                mean_name = i[1],
                                                                alpha = K_alpha,
                                                                use_anomaly = i[2])

                    # заполним X_train 
                    sub_test_ind = list(X_test_sub.index)
                    X_train.loc[sub_test_ind] = X_test_sub
            make_feat +=1
            print('Создано',make_feat,'фичей из ',n_col)


        ################################

        # save
        if mode == 'validation':
            target_train_buf.append(y_train)
            target_test_buf.append(y_test)
            X_train.to_pickle(str(num_iter) + "X_train_Fold.pkl")
            X_test.to_pickle(str(num_iter) + "X_test_Fold.pkl")
            y_train.to_pickle(str(num_iter) + "y_train_Fold.pkl")
            y_test.to_pickle(str(num_iter) + "y_test_Fold.pkl")
          
        ################################


        X_train_to_model = X_train.drop(params_dict['To_mean_encode']+
                                        ['speed','is_anomaly']+
                                        ['Time from Pickup to Arrival'],axis=1)
        if mode == 'validation':
            X_test_to_model = X_test.drop(params_dict['To_mean_encode']+
                                          ['speed','is_anomaly']+
                                          ['Time from Pickup to Arrival','help_feature'],axis=1)
        
        if mode == 'prediction':
            X_test_to_model = X_test.drop(params_dict['To_mean_encode']+['help_feature'],axis=1)
        
        print('ШЕЙПЫ ТРЕЙНА ТЕСТА',X_train_to_model.shape[1], X_test_to_model.shape[1])

#         print(X_train_to_model.columns)
#         print(X_test_to_model.columns)
        assert X_train_to_model.shape[1] == X_test_to_model.shape[1], 'Fail'


        categorical_features_indices = np.where(X_train_to_model.dtypes != np.float)[0] ### !!!!
        if mode == 'validation':
            clf.fit(X_train_to_model,y_train,eval_set=(X_test_to_model,y_test),cat_features = categorical_features_indices )
        if mode == 'prediction':
             clf.fit(X_train_to_model,y_train,cat_features = categorical_features_indices )
                
        preds = clf.predict(X_test_to_model)
        
        if mode == 'validation':
            err = np.sqrt(mse(y_test, preds))
            err_buf.append(err)
            print('RMSE = ' + str(err))
            preds_buf.append(preds)
            print('Mean RMSE = ' + str(np.mean(err_buf)) + ' +/- ' + str(np.std(err_buf)))
            num_iter+=1
        if mode == 'prediction':
            return X_train_to_model,X_test_to_model,preds
        
    return X_train_to_model,X_test_to_model,preds_buf,target_train_buf,target_test_buf,err_buf
```

```python
MODE = 'validation'
ALPHA = 5
SUBSET = 10
X_train_to_model,X_test_to_model,preds_buf,target_train_buf,target_test_buf,err_buf = test_prediction(X,y,K_alpha=ALPHA,mode = MODE,n_splits_sub=SUBSET)
```
